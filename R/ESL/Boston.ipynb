安装 GGaly

```r
if (!require(GGally)) {
  install.packages("GGally")  # 如果未安装，则安装
  library(GGally)  # 加载包
} else {
  print("GGally 已安装")
}
```

加载Boston 数据集

library(MASS)
data(Boston)

```r
colnames(Boston)
```

```r
# 简单的散点图
plot(Boston$black, Boston$crim,
     xlab = "黑人比例", ylab = "犯罪率",
     main = "黑人比例与犯罪率的散点图")
```

```r
# 绘制散点图
ggplot(Boston, aes(x = black, y = crim)) +
  geom_point() +
  labs(x = "黑人比例", y = "犯罪率", title = "黑人比例与犯罪率的散点图") +
  theme_minimal()
```

```r
truehist(Boston$crim, main = "犯罪率的直方图")
```
